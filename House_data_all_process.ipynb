```python
#imputing missing values with sicketlearn
#Divide original data independent & dependent variables 
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
# data divied x & y
x=df.drop(columns=["SalePrice"])
y=df["SalePrice"]
```

```python
x
```

```python
y
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=None)
```

```python
#shape check
print("Shape",x_train.shape,y_train.shape)
print("Shape",x_test.shape,y_test.shape)
```

```python
imputer1=SimpleImputer(strategy="mean")
imputer2=SimpleImputer(strategy="mean")
imputer3=SimpleImputer(strategy="mean")
```

```python
from numpy.ma.core import remainder
trf=ColumnTransformer([
                       ("imputer1",imputer1,["MasVnrArea"]),
                       ('imputer2',imputer2,["GarageYrBlt"]),
                       ("imputer3",imputer3,["LotFrontage"])
],remainder="passthrough")
```

```python
trf.fit(x_train)
```

```python
x_train.isnull().sum()
```

```python
x_train_trf=trf.transform(x_train)
```

```python
x_train_trf
```

```python
x_test_trf=trf.transform(x_test)
```

```python
x_test_trf
```

```python
x_train=pd.DataFrame(x_train_trf,columns=x_train.columns)
x_test=pd.DataFrame(x_test_trf,columns=x_test.columns)
```

```python
x_train.isnull().sum().sort_values(ascending=False)
```

```python
x_test.isnull().sum().sort_values(ascending=False)
```

```python
imputer=SimpleImputer(strategy="most_frequent")
```

```python
x_train_ob_trf=imputer.fit_transform(x_train)
x_test_ob_trf=imputer.fit_transform(x_test)
```

```python
x_test=pd.DataFrame(x_test_ob_trf,columns=x_test.columns)
x_train=pd.DataFrame(x_train_ob_trf,columns=x_train.columns)
```
